# Data Visualization

### Import necessary libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

### Read data and store in football_data variable



```python
football_data = pd.read_csv('../football_data.csv')
football_data.fillna(method ='ffill', inplace = True)
```

### Cleaning up data



```python
football_data['Position'] = football_data['Position'].astype(str)
```

### Function that takes parameters and shows histogram using matplotlib.pyplot

```python
def plotHistogram (data, bins, xlabel, ylabel, title) :
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.hist(x = data, bins = bins)
    plt.show()
```

### Plotting histograms for position, age and skill moves

```python
# Plot positions
plotHistogram(football_data['Position'], 28, "Position", "Frequency", "Player Positions Histogram")
# Plot ages
plotHistogram(football_data['Age'], range(15, 46, 1), "Age", "Frequency", "Player Age Histogram")
# Plot skill moves
plotHistogram(football_data['Skill Moves'], range(1,6,1), "Skill Moves", "Frequency", "Player Skill Moves Histogram")
```

### Finding the most common clubs and countries

```python
countries = football_data['Nationality'].value_counts()[:5].index.tolist()
clubs = football_data['Club'].value_counts()[:5].index.tolist()
positions = football_data['Position'].value_counts()[:5].index.tolist()
colors = ['g', 'r', 'y', 'm', 'c']
```

### Plot distribution of ages across most common countries

```python
for idx, country in enumerate(countries):

    plt.hist(football_data.loc[football_data['Nationality'] == country]['Age'], color = colors[idx], label = country, alpha = 0.35)

plt.xlabel('Age')    
plt.ylabel('Frequency')    
plt.legend()
plt.show()
```

### Plot distribution of ages across most common clubs

```python
for idx, club in enumerate(clubs):

    plt.hist(football_data.loc[football_data['Club'] == club]['Age'], color = colors[idx], label = club, alpha = 0.35)

    
plt.xlabel('Age')    
plt.ylabel('Frequency')    
plt.legend()
plt.show()
plt.show()
```

### Plot distribution of heights across most common countries

```python
for idx, country in enumerate(countries):

    plt.hist(football_data.loc[football_data['Nationality'] == country]['Height'], color = colors[idx], label = country, alpha = 0.35)

plt.xlabel('Height')    
plt.ylabel('Frequency')    
plt.legend()
plt.show()
    
plt.show()
```

### Plot distribution of heights across most common clubs

```python
for idx, club in enumerate(clubs):

    plt.hist(football_data.loc[football_data['Club'] == club]['Height'], color = colors[idx], label = club, alpha = 0.35)

    
plt.xlabel('Height')    
plt.ylabel('Frequency')    
plt.legend()
plt.show()
plt.show()
```

### Plot distribution of weights across most common countries

```python
for idx, country in enumerate(countries):

    plt.hist(football_data.loc[football_data['Nationality'] == country]['Weight'], color = colors[idx], label = country, alpha = 0.35)
    

plt.xlabel('Weight')    
plt.ylabel('Frequency')    
plt.legend()    
plt.show()
```

### Plot distribution of weights across most common clubs

```python
for idx, club in enumerate(clubs):

    plt.hist(football_data.loc[football_data['Club'] == club]['Weight'], color = colors[idx], label = club, alpha = 0.35)
 
plt.xlabel('Weight')    
plt.ylabel('Frequency')    
plt.legend()    
plt.show()
```

```python
sns.lmplot(x='Overall', y='Potential', hue='Nationality', fit_reg=False, \
    data=football_data.loc[football_data['Nationality'].isin(['Argentina', 'Brazil', 'Portugal', 'Chile'])])
```

### Age of players according to position
